```python
"""
Podemos observar que los titulos más comunes son Spiderman, batman.
Hay muchísimas ediciones de cada comic, que habrá que analizar mucho más a fondo.
El precio por comic más visible es 3.99, aunque se salen de lo comun, los gratuitos y los objetos de coleccionista.
La mayoría de comics tienen una tirada baja, habrá que desglosarlo un poco más.
"""
```

```python
# Convertimos a texto y limpiamos
df1['cover_price'] = df1['cover_price'].astype(str).replace(["Free", r'[\$,]'], '', regex=True)
df1['cover_price'] = pd.to_numeric(df1['cover_price'], errors='coerce')

df1['current_value'] = df1['current_value'].astype(str).replace(r'[\$,]', '', regex=True)
df1['current_value'] = pd.to_numeric(df1['current_value'], errors='coerce')

# Media y desviación típica
print(f"Precio de portada medio: {df1['cover_price'].mean():.2f} $")
print(f"Valor actual medio: {df1['current_value'].mean():.2f} $")

# Coeficiente de variación como indicador de dispersión
cv = df1['current_value'].std() / df1['current_value'].mean()
print(f"Coeficiente de variación del valor actual: {cv:.2f}")
```

```python
# Podemos ver que hay algunos comics, que se ha revalorizado, algunos con unos cambios drásticos del precio, y otros no tanto.
```

```python
# Extraer año del campo cover_date si hace falta
df1['cover_year'] = df1['cover_date'].str.extract(r"(\d{2,4})$")

# Convertir a numérico y analizar distribución
df1['cover_year'] = pd.to_numeric(df1['cover_year'], errors='coerce')
print(df1['cover_year'].value_counts().sort_index())
```

```python
df1['cover_year'] = df1['cover_date'].str.extract(r'(19\d{2}|20[0-2]\d)').astype(float)
```

```python
# Convertir a numérico y analizar distribución
df1['cover_year'] = pd.to_numeric(df1['cover_year'], errors='coerce')
print(df1['cover_year'].value_counts().sort_index())
```

```python
df1['decade'] = (df1['cover_year'] // 10) * 10

# Contar cómics por década
conteo_decadas = df1['decade'].value_counts().sort_index()
plt.figure(figsize=(10, 5))
conteo_decadas.plot(kind='bar', color='darkorange')
plt.title('Número de cómics por década de publicación')
plt.xlabel('Década')
plt.ylabel('Número de cómics')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Top 10 cómics más buscados
print(df1[['title', 'issue', 'searched']].sort_values(by='searched', ascending=False).head(10))
```

```python
# Editoriales coin más titulos

print(df1['pub_name'].value_counts().head(10))
```

```python
# Porcentaje de cómics sin autores/personajes asignados
sin_contribuidores = df1['contributors_names'].str.contains('no contributors assigned', na=True).mean()
sin_personajes = df1['characters'].str.contains('no characters assigned', na=True).mean()

print(f"Cómics sin contribuidores: {sin_contribuidores:.2%}")
print(f"Cómics sin personajes: {sin_personajes:.2%}")
```

```python
# Rellenar valores nulos según el tipo de columna:
# Si es numérica: con la mediana (menos sensible a outliers)
# Si es categórica: con la moda (el valor más común)
for col in df1.columns:
    if df1[col].dtype in ['float64', 'int64']:
        df1[col] = df1[col].fillna(df1[col].median())
    else:
        df1[col] = df1[col].fillna(df1[col].mode()[0])

print(df1.isna().sum().sum())  # Debe dar 0
```

```python
# Distribución del valor actual
sns.histplot(df1["current_value"], bins=30, kde=True)
plt.title("Distribución del valor actual del cómic")
plt.xlabel("Valor actual")
plt.ylabel("Frecuencia")
plt.show()

# Distribución por editorial
sns.countplot(data=df1, x="pub_name", order=df1["pub_name"].value_counts().index[:10])
plt.xticks(rotation=45)
plt.title("Top 10 editoriales por número de cómics")
plt.show()
```

```python
# Relación entre el año de publicación y el valor actual
sns.scatterplot(data=df1, x="cover_year", y="current_value")
plt.title("Relación entre año de publicación y valor actual")
plt.xlabel("Año de publicación")
plt.ylabel("Valor actual")
plt.show()

# Relación entre precio original y valor actual
sns.scatterplot(data=df1, x="cover_price", y="current_value")
plt.title("¿El precio original influye en el valor actual?")
plt.xlabel("Precio de portada")
plt.ylabel("Valor actual")
plt.show()

# Boxplot del valor actual según década
df1["decade"] = (df1["cover_year"] // 10) * 10
sns.boxplot(data=df1, x="decade", y="current_value")
plt.title("Valor actual por década de publicación")
plt.xlabel("Década")
plt.ylabel("Valor actual")
plt.show()
```

```python
# Correlación entre variables numéricas
numeric_cols = df1.select_dtypes(include=np.number)
corr = numeric_cols.corr()

# Mapa de calor
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de correlación")
plt.show()
```

```python
# Calcular correlación de todas las variables con "current_value"
correlaciones = df1.corr(numeric_only=True)["current_value"].sort_values()

# Visualizarlo
correlaciones.drop("current_value").plot(kind="barh", figsize=(8,6))
plt.title("Correlación con current_value")
plt.xlabel("Coeficiente de correlación")
plt.show()
```

```python
import scipy.stats as stats

# Correlación de Pearson entre 'cover_price' y 'current_value'
corr, p_value = stats.pearsonr(df1['cover_price'], df1['current_value'])

print(f"Correlación de Pearson: {corr}")
print(f"P-valor: {p_value}")

# Estas dos variables están directamente relacionadas
```

```python
variables_numericas = ['cover_price', 'cover_year', 'issues_total', 'pub_issues_total']  # Añade más si tienes
resultados = []

for var in variables_numericas:
    corr, p = stats.pearsonr(df1[var], df1['current_value'])
    resultados.append((var, corr, p))

# Mostrar resultados ordenados por correlación absoluta
resultados_df1 = pd.DataFrame(resultados, columns=['Variable', 'Correlación de Pearson', 'P-valor'])
resultados_df1['Abs Correlación'] = resultados_df1['Correlación de Pearson'].abs()
resultados_df1.sort_values('Abs Correlación', ascending=False, inplace=True)

print(resultados_df1)
```
