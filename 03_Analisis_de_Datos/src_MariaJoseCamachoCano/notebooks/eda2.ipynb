```python
from scipy.stats import f_oneway

conteo_editoriales = df1['pub_name'].value_counts()
top_editoriales = conteo_editoriales[conteo_editoriales > 10].index
df_anova = df1[df1['pub_name'].isin(top_editoriales)].dropna(subset=['current_value'])
# Agrupamos precios por editorial
precios_por_editorial = [group['current_value'].values for name, group in df_anova.groupby('pub_name')]

# Test ANOVA
stat, pval = f_oneway(*precios_por_editorial)
print(f"p-valor ANOVA: {pval:.4f}")
```

```python
# 1. Preparar datos: variables predictoras y target
df_reg = df1.dropna(subset=['year_start','issues_total','cover_price','current_value'])
X = df_reg[['year_start','issues_total','cover_price']].values
y = df_reg['current_value'].values

# 2. Añadir término independiente (bias)
X_design = np.hstack([np.ones((X.shape[0],1)), X])

# 3. Calcular coeficientes con mínimos cuadrados
beta, *_ = np.linalg.lstsq(X_design, y, rcond=None)

# 4. Crear un DataFrame con coeficientes
coef_names = ['Intercept','Year','Issues_total','Cover_price']
df_coef = pd.DataFrame({'Variable': coef_names, 'Coeficiente': beta})

# 5. Visualizar coeficientes
plt.figure(figsize=(6,4))
plt.bar(df_coef['Variable'], df_coef['Coeficiente'], color='steelblue')
plt.title('Coeficientes de Regresión Lineal Múltiple')
plt.ylabel('Valor del Coeficiente')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 1. Filtrar y discretizar ‘issues_total’ en 5 categorías
df_hm = df1.dropna(subset=['year_start','issues_total','current_value']).copy()
df_hm['issues_cat'] = pd.qcut(df_hm['issues_total'], 5, labels=False)

# 2. Pivot table: precio medio para cada (year_start, issues_cat)
pivot = df_hm.pivot_table(
    index='year_start',
    columns='issues_cat',
    values='current_value',
    aggfunc='mean'
)

# 3. Dibujar heatmap
plt.figure(figsize=(12,8))
sns.heatmap(
    pivot,
    cmap='YlGnBu',
    linewidths=0.5,
    linecolor='white',
    cbar_kws={'label':'Precio Medio (€)'}
)
plt.title('Precio Medio por Año de Publicación y Escasez (categorías)')
plt.xlabel('Categoría de Nº de Ejemplares')
plt.ylabel('Año de Publicación')
plt.tight_layout()
plt.show()
```

```python
# 1. Calcular valores ajustados y residuos
y_pred = X_design.dot(beta)
residuos = y - y_pred

# 2. Gráfico residuos vs. predicción
plt.figure(figsize=(8,5))
plt.scatter(y_pred, residuos, alpha=0.3, s=20)
plt.axhline(0, color='red', linestyle='--')
plt.title('Diagnóstico de Residuos')
plt.xlabel('Precio Ajustado (€)')
plt.ylabel('Residuo (€)')
plt.tight_layout()
plt.show()
```

```python
# Chi2 entre precio y rareza

# 1. Eliminar nulos
df_chi = df1.dropna(subset=['current_value', 'issues_total']).copy()

# 2. Categorizar el precio en terciles
df_chi['precio_cat'] = pd.qcut(df_chi['current_value'], q=3, labels=['Bajo', 'Medio', 'Alto'])

# 3. Categorizar 'issues_total' en 3 grupos (de escaso a común)
df_chi['issues_cat'] = pd.qcut(df_chi['issues_total'], q=3, labels=['Escaso', 'Moderado', 'Común'])

# 4. Tabla de contingencia
tabla = pd.crosstab(df_chi['precio_cat'], df_chi['issues_cat'])

# 5. Test Chi²
chi2, p, dof, expected = chi2_contingency(tabla)

print("Chi²:", chi2)
print("p-valor:", p)
```

```python
# chi2 entre precio y decada de publicacion
# 1. Eliminar nulos
df_chi2 = df1.dropna(subset=['current_value', 'year_start']).copy()

# 2. Categorizar precio en terciles
df_chi2['precio_cat'] = pd.qcut(df_chi2['current_value'], q=3, labels=['Bajo', 'Medio', 'Alto'])

# 3. Crear variable de década
df_chi2['decada'] = (df_chi2['year_start'] // 10) * 10
df_chi2['decada'] = df_chi2['decada'].astype(int).astype(str) + 's'  # Ej: "1980s"

# 4. Tabla de contingencia
tabla_decada = pd.crosstab(df_chi2['precio_cat'], df_chi2['decada'])

# 5. Test Chi²
chi2, p, dof, expected = chi2_contingency(tabla_decada)

print("Chi²:", chi2)
print("p-valor:", p)
```

```python

# Variables candidatas a analizar
variables_categoricas = ['rating', 'owned', 'searched', 'pages', 'pub_name', 
                         'pub_titles_total', 'pub_issues_total', 'year_start', 'decade']

# Copia del DataFrame con nulos eliminados en precio
df_chi = df1.dropna(subset=['current_value']).copy()
df_chi['precio_cat'] = pd.qcut(df_chi['current_value'], q=3, labels=['Bajo', 'Medio', 'Alto'])

# Función auxiliar para categorizar variables numéricas si es necesario
def categorizar_si_necesario(df, col):
    if pd.api.types.is_numeric_dtype(df[col]):
        nueva_col = col + '_cat'
        df[nueva_col] = pd.qcut(df[col], q=3, duplicates='drop')
        return nueva_col
    return col

# Ejecutar test Chi² para cada variable
resultados = []
for var in variables_categoricas:
    try:
        var_cat = categorizar_si_necesario(df_chi, var)
        tabla = pd.crosstab(df_chi['precio_cat'], df_chi[var_cat])
        chi2, p, _, _ = chi2_contingency(tabla)
        resultados.append({'Variable': var, 'p-valor': p})
    except Exception as e:
        resultados.append({'Variable': var, 'p-valor': f'Error: {e}'})

# Crear DataFrame y filtrar los que tienen p-valor numérico
df_resultados = pd.DataFrame(resultados)

# Función para identificar si un valor es numérico
def es_numerico(x):
    return isinstance(x, (int, float, np.number))

df_validos = df_resultados[df_resultados['p-valor'].apply(es_numerico)]
df_validos = df_validos.sort_values(by='p-valor')

# Mostrar resultados
print(df_validos)
```
